```python
#40. Crie um programa que leia 5 notas usando for e calcule a média da turma.

soma = 0 # Minha caixinha começa vazia

# O loop vai rodar 5 vezes
for i in range(1, 6):
    nota = float(input(f"Digite a {i}ª nota: "))
    soma = soma + nota # Pega o que já tem na caixa e soma a nova nota

# Depois que o loop acaba, calculamos a média
media = soma / 5

print(f"A média da turma é: {media:.1f}")
```

```python
#41. Crie um programa que peça a idade de várias pessoas até que seja digitado 999. Ao final, mostre quantas são maiores de idade.

maiores = 0
idade = 0

# Pedimos a primeira idade para o loop poder começar
idade = int(input("Digite a idade (ou 999 para parar): "))

while idade != 999:
    if idade >= 18:
        maiores += 1
    
    # Pedimos a próxima idade dentro do loop
    idade = int(input("Digite a idade (ou 999 para parar): "))

print(f"Total de pessoas maiores de idade: {maiores}")
```

```python
#42. Crie um programa que peça números até que seja digitado um número negativo. Ao final, mostre o maior número digitado.

maior = 0
numero = 0

# Pedimos o primeiro número
numero = float(input("Digite um número (negativo para parar): "))

# O loop roda enquanto o número for positivo ou zero
while numero >= 0:
    # Se o número atual for maior que o 'maior' que temos guardado
    if numero > maior:
        maior = numero
    
    # Pedimos o próximo número dentro do loop
    numero = float(input("Digite um número (negativo para parar): "))

print(f"O maior número digitado foi: {maior}")
```

```python
#43. Crie um programa que implemente um menu com as opções: 1 - Somar, 2 - Subtrair, 3 - Sair. O menu deve repetir até o usuário escolher sair.

opcao = 0

# O menu repete ENQUANTO a opção não for 3
while opcao != 3:
    print("\n--- MENU ---")
    print("1 - Somar")
    print("2 - Subtrair")
    print("3 - Sair")
    
    opcao = int(input("Escolha uma opção: "))
    
    if opcao == 1:
        n1 = float(input("Digite o primeiro número: "))
        n2 = float(input("Digite o segundo número: "))
        print(f"Resultado da soma: {n1 + n2}")
        
    elif opcao == 2:
        n1 = float(input("Digite o primeiro número: "))
        n2 = float(input("Digite o segundo número: "))
        print(f"Resultado da subtração: {n1 - n2}")
        
    elif opcao == 3:
        print("Saindo do programa... Até logo!")
        
    else:
        print("Opção inválida! Tente novamente.")
```

```python
#44. Crie um programa que use while para validar uma nota entre 0 e 10. Enquanto a nota for inválida, peça novamente.

# Pedimos a nota pela primeira vez
nota = float(input("Digite uma nota entre 0 e 10: "))

# ENQUANTO a nota for menor que 0 OU maior que 10
while nota < 0 or nota > 10:
    print("Valor inválido!")
    # Pede a nota novamente para tentar sair do loop
    nota = float(input("Digite uma nota válida (entre 0 e 10): "))

print(f"Nota {nota} validada com sucesso!")
```

```python
#45. Crie um programa que peça o nome de produtos e seus preços até que o usuário digite 'fim'. Ao final, mostre o total da compra.

total = 0
nome_produto = ""

# Enquanto o nome do produto for diferente de 'fim'
while nome_produto != "fim":
    nome_produto = input("Digite o nome do produto (ou 'fim' para encerrar): ").lower()
    
    if nome_produto != "fim":
        preco = float(input(f"Digite o preço de {nome_produto}: "))
        total += preco # Acumulando o valor

print(f"\nCompra finalizada! O total foi: R$ {total:.2f}")
```
